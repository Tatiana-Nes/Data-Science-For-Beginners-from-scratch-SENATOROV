1.  Что делает команда python -m venv venv?
    Создает виртуальное окружение в папке venv

    1.1 Что делает каждая команда в списке ниже?
        pip list - Показывает установленные в текущем окружении пакеты и их версии в виде таблицы.
        pip freeze > requirements.txt - выгружает зависимости в файл requirements.txt
        pip install -r requirements.txt - устанавливает зависимости из файл в файл requirements.txt

2.  Что делает каждая команда в списке ниже?
    conda env list - показывает все среды Conda и путь к ним
    conda create -n env_name python=3.5 - создаёт новое виртуальное окружение с именем env_name и установленным Python 3.5. 
    conda env update -n env_name -f file.yml - Обновляет виртуальное окружение env_name (Добавляет/обновляет пакеты) по спецификации из YAML.
    source activate env_name - активирует виртуальное окружение env_name
    source deactivate - деактивирует текущее виртуальное окружение
    conda clean -a - чистит кеш

4.  Как установить необходимые пакеты внутрь виртуального окружения для conda/venv?
    Чтоьбы установить пакеты внутрь окружения через conda нужно выполнить команду:
    conda install pandas numpy

    Чтоьбы установить пакеты внутрь venv:
    python -m pip install --upgrade pip
    pip install pandas numpy

11. Зачем нужно виртуально окружение?
    Для того, чтобы при работе над разными проектами, где нам необходимы разные пакеты, избежать конфликтов библиотек
    Все пакеты лежат внутри отдельной папки, в случае чего можно легко удалить окружение, не трогая систему.
